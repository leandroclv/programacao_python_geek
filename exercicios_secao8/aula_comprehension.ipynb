###List Comprehension

- Utilizando list comprehension nõs podemos gerar novas listas com dados processados a partir de outro iterável

Sintaxe da list comprehension
- [dado for dado in iterável]

```python
#Para entender melhor o acontecerá, devemos dividir a expressão em duas partes: a primeira parte: for numero in numeros/ a segunda parte: numero * 10 
numeros = [1,2,3,4,5]
res = [numero * 10 for numero in numeros]
print(res)
```

```python
res = [numero / 2 for numero in numeros]
print(res)
```

```python
def funcao(valor):
    return valor * valor

res = [funcao(numero) for numero in numeros]
print(res)
```

```python
#List comprehension vs loop

#Loop
numeros = [1,2,3,4,5]
numeros_dobrados = []

for numero in numeros:
    numero_dobrados = numero * 2
    numeros_dobrados.append(numero_dobrados)
    
print(numeros_dobrados)
```

```python
numeros_dobrados = [numero * 2 for numero in numeros]
print(numeros_dobrados)
```

```python
nome = 'Geek University'
print([letra.upper() for letra in nome])
```

```python
def caixa_alta(nome):
    nome = nome.replace(nome[0], nome[0].upper())
    return nome

amigos = ['maria', 'júlia', 'pedro', 'guilherme', 'vanessa']
print([caixa_alta(amigo) for amigo in amigos])
```

```python
print([numero * 3 for numero in range(1,10)])
```

```python
print([bool(valor) for valor in [0,[], '', True,1, 3.14]])
```

```python
print([str(numero) for numero in [1,2,3,4,5]])
```

```python
#Nós podemos adicionar estruturas condicionais lógicas as nossas List Comprehension

numeros = [1,2,3,4,5,6]

pares = [numero for numero in numeros if numero % 2 == 0]
pares
```

```python
impares = [numero for numero in numeros if numero % 2 != 0]
impares
```

```python
#Refatorar

#Qualquer número par por módulo de 2 é 0 e o é False em Python. not False = True
pares = [numero for numero in numeros if not numero % 2]
pares
```

```python
#Qualquer número ímpar pormódulo de 2 é 1, e 1 é True em Python
impar = [numero for numero in numeros if numero % 2]
impar
```

```python
res = [numero * 2 if numero % 2 == 0 else numero/2 for numero in numeros]
res
```

```python
#Listas aninhadas
listas = [[1,2,3], [4,5,6], [7,8,9]] #Matriz 3 X 3

print(listas[0][1])
print(listas[2][-2])
```

```python
#Iterando com loops em uma lista aninhadas
for lista in listas:
    for numero in lista:
        print(numero, end=' ')
```

```python
#Utilizando list comprehension

[[print(valor) for valor in lista] for lista in listas]
```

```python
#Gerando uma matriz 3x3

tabuleiro = [[numero  for numero in range(1,4)] for valor in range(1,4)]
tabuleiro
```

```python
#Gerando jogadas para o jogo da velha
velha = [['x' if numero % 2 == 0 else '0' for numero in range(1,4)] for valor in range(1,4)]
velha
```

```python
#Gerando valor inicial
print([['*' for i in range(1,4)] for j in range(1,4)])
```
