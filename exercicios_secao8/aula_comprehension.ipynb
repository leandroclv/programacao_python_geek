###List Comprehension

- Utilizando list comprehension nõs podemos gerar novas listas com dados processados a partir de outro iterável

Sintaxe da list comprehension
- [dado for dado in iterável]

```python
#Para entender melhor o acontecerá, devemos dividir a expressão em duas partes: a primeira parte: for numero in numeros/ a segunda parte: numero * 10 
numeros = [1,2,3,4,5]
res = [numero * 10 for numero in numeros]
print(res)
```

```python
res = [numero / 2 for numero in numeros]
print(res)
```

```python
def funcao(valor):
    return valor * valor

res = [funcao(numero) for numero in numeros]
print(res)
```

```python
#List comprehension vs loop

#Loop
numeros = [1,2,3,4,5]
numeros_dobrados = []

for numero in numeros:
    numero_dobrados = numero * 2
    numeros_dobrados.append(numero_dobrados)
    
print(numeros_dobrados)
```

```python
numeros_dobrados = [numero * 2 for numero in numeros]
print(numeros_dobrados)
```

```python
nome = 'Geek University'
print([letra.upper() for letra in nome])
```

```python
def caixa_alta(nome):
    nome = nome.replace(nome[0], nome[0].upper())
    return nome

amigos = ['maria', 'júlia', 'pedro', 'guilherme', 'vanessa']
print([caixa_alta(amigo) for amigo in amigos])
```

```python
print([numero * 3 for numero in range(1,10)])
```

```python
print([bool(valor) for valor in [0,[], '', True,1, 3.14]])
```

```python
print([str(numero) for numero in [1,2,3,4,5]])
```
