```python
#Função em Python
def funcao(x):
    return 3 * x + 1

print(funcao(4))
print(funcao(7))
```

```python
#Expressão Lambda
funcao = lambda x: 3 * x + 1

print(funcao(4))
print(funcao(7))
```

```python
#Podemos ter expressões lambdas com multiplas entradas
nome_completo = lambda nome, sobrenomes: f'{nome.strip().title()} {sobrenomes.strip().title()}'

print(nome_completo('leandro', 'costa lima verde'))
```

```python
#Em funções Python podemos ter nenhuma ou várias entradas. Em lambda também 
amar = lambda: 'Como não amaer Python?'

uma = lambda x: 3 * x + 1

duas = lambda x, y: (x - y) ** 0.5

tres = lambda x, y, z: 3 / (1 / x*1 /y*1 /z)

#n = lambda x1, x1, ... xn: expressao

#OBS: Se passarmos mais argumentos do que parâmetros esperados teremos TypeErro
```

```python
print(amar())
print(uma(6))
print(duas(5,7))
print(tres(3,6,9))
```

```python
autores = ['Isaac Asimov', 'Ray Bradbury', 'Robert Heinlein', 'Arthur C. Clarke', 'Frank Verbert', "Orson Scott Card", 'Douglas Adams', 'M. G. Wells', 'Leigh Brackett']

autores.sort(key=lambda sobrenome: sobrenome.split(' ')[-1].lower())
print(autores)
```

```python
def geradora_funcao_quadratica(a,b,c):
    return lambda x: a * x ** 2 + b * x + c

teste = geradora_funcao_quadratica(2,3,-5)

print(teste(0))
print(teste(1))
print(teste(2))

#ou

print(geradora_funcao_quadratica(3,0,1)(2))
```
