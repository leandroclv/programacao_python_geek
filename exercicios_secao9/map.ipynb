```python
import math
```

```python
def area(r):
    return math.pi * (r ** 2)


print(area(2))
print(area(5.3))
```

```python
raios = [2,5,7.1,0.3,10,44]

#Forma comum
areas = []
for r in raios:
    areas.append(area(r))

print(areas)
```

```python
#Forma com map
#Map é uma função que recebe dois parâmetros: o primeiro a função (que deve receber somente um parâmetro) e o segundo um iterável. 

areas = map(area, raios)
#retorna um <map object at 0x76412c134e80>, por isso temos que converter para lista ou tupla
print(list(areas))
```

```python
numbers = [1,2,3,4,5,6]
soma = map(lambda x: x + x, numbers)
print(list(soma))
```

```python
print(list(map(lambda r: math.pi * (r ** 2), raios)))
```

```python
#OBS: Após utilizar a função map() depois da primeira utilização do resultado, ele zera
```

```python
cidades = [('Berlim', 29), ('Cairo', 36), ('Buenos Aires', 19), ('Los Angeles', 26), ('Tokio', 27), ('Nova York', 28), ('Londres', 22)]

c_to_f = lambda dado: (dado[0], (9/5) * dado[1] + 32)

print(list(map(c_to_f, cidades)))
```
